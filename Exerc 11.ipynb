```python
from math import sqrt

print('Equação de segundo grau')
print(' ')
print('ax² + bx + c')

fim = False
while not fim:

    try:
        a = float(input('Insira o coeficiente "a" da equação: ax² '))
        fim = True
    except:
        print('Inserir um número válido')
    

    fim = False     

    while not fim:        
        try:    
            b = float(input('Insira o coeficiente "b" da equação: bx '))
            fim = True
        except:
            print('Inserir um número válido')
            
    fim = False
    
    while not fim:
        
        try:    
            c = float(input('Insira o coeficiente "c" da equação: '))
            fim = True
        except:
            print('Inserir um número válido')
    
program = True

while program:
    if a == 0:
        print('Esta equação não é do segundo grau!')
        program = False
        break
    
       
    delta = (b)**2 - 4 * a * c
    print('Delta = ', delta)
    print(' ')
    print(' ')

    if delta < 0:
        print('Delta negativo!')
        print('Esta equação de segundo grau não possui raízes reais.')
        program = False
        break
        
    
    if delta == 0:
        print('Delta igual a zero!')
        print(' ')
        print(' ')
        print('A equação possui apenas uma raiz real: ')
        print(' ')
        print(' ')
        x = -b / (2*a)
        print('x = ', round(x,2))
        program = False
    
    else:  
        print('Delta Positivo!')
        print(' ')
        print(' ')
        print('A equação possui duas raízes reais: ')
        print(' ')
        print(' ')
        x_1 = (-b + sqrt(delta) ) / (2*a)
        x_2 = (-b - sqrt(delta) ) / (2*a)
        print('x1 = ', round(x_1,2))
        print('x2 = ', round(x_2,2))
        program = False
```
